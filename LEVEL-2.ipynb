# LEVEL 2 : TASK 3

Task: Geographic Analysis

Q1] Plot the locations of restaurants on a map using longitude and latitude coordinates.

```python
import plotly.graph_objects as go
# Create a scatter mapbox plot
fig = go.Figure(go.Scattermapbox(
    lat=df['Latitude'],
    lon=df['Longitude'],
    mode='markers',
    marker=go.Marker(
        size=8,
        color='red',
        opacity=0.8
    )
))

# Set the mapbox style and center
fig.update_layout(
    mapbox_style="open-street-map",
    mapbox_center={"lat": df['Latitude'].mean(), "lon": df['Longitude'].mean()},
    mapbox_zoom=5
)

# Show the plot
fig.show()
```

Q2] Identify any patterns or clusters of restaurants in specific areas.

```python
import plotly.graph_objs as go
import scipy.cluster.hierarchy as hierarchy

# Create a scatter mapbox plot
fig = go.Figure(go.Scattermapbox(
    lat=df['Latitude'],
    lon=df['Longitude'],
    mode='markers',
    marker=go.Marker(
        size=8,
        color='red',
        opacity=0.8
    )
))

# Apply clustering algorithm
Z = hierarchy.linkage(df[['Latitude', 'Longitude']], method='ward')
criteria = hierarchy.fcluster(Z, t=0.05*max(df['Latitude']), criterion='distance')

# Add cluster markers to the plot
for i in range(len(criteria)):
    if criteria[i] == i:
        cluster_df = df.loc[criteria == i]
        fig.add_trace(go.Scattermapbox(
            lat=[cluster_df['Latitude'].mean()]*len(cluster_df),
            lon=[cluster_df['Longitude'].mean()]*len(cluster_df),
            mode='markers',
            marker=go.Marker(
                size=20,
                color='blue',
                opacity=0.8
            )
        ))

# Set the mapbox style and center
fig.update_layout(
    mapbox_style="open-street-map",
    mapbox_center={"lat": df['Latitude'].mean(), "lon": df['Longitude'].mean()},
    mapbox_zoom=5
)

# Show the plot
fig.show()
```

# LEVEL 2 : TASK 4

Task: Restaurant Chains

Q1] Identify if there are any restaurant chains present in the dataset.

```python
# Extract chain names from the 'Restaurant Name' column
chains = df['Restaurant Name'].apply(lambda x: x.split()[0]).unique()

if len(chains) > 1:
    print("Restaurant chains present in the dataset:")
    for chain in chains:
        print(chain)
else:
    print("No restaurant chains present in the dataset.")
```

Q2]Analyze the ratings and popularity of different restaurant chains.

```python
# Calculate the average rating for each restaurant chain
avg_rating_per_chain = df.groupby('Restaurant Name')['Aggregate rating'].mean()

# Calculate the popularity of each restaurant chain based on the total number of votes
popularity_per_chain = df.groupby('Restaurant Name')['Votes'].sum()

# Create a DataFrame to combine average rating and popularity
chain_analysis = pd.DataFrame({
    'Average Rating': avg_rating_per_chain,
    'Total Votes': popularity_per_chain
})

sorted_chain_analysis = chain_analysis.sort_values(by='Average Rating', ascending=False)

print(sorted_chain_analysis)
```
